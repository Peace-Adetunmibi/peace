```python
# Toy dataset: 1D positions -> energy
# Energy = x^2 (just for demonstration)
X = torch.tensor([[0.0], [1.0], [2.0], [3.0]], dtype=torch.float32)
Y = torch.tensor([[0.0], [1.0], [4.0], [9.0]], dtype=torch.float32)
```

```python
model = nn.Sequential(
    nn.Linear(1, 10),   # Input layer -> 10 hidden neurons
    nn.ReLU(),          # Activation function
    nn.Linear(10, 1)    # Output layer -> predicts energy
)
```

```python
criterion = nn.MSELoss()                 # Mean Squared Error loss
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)  # Adam optimizer
```

```python
# Training loop
for epoch in range(500):
    optimizer.zero_grad()      # Reset gradients
    y_pred = model(X)          # Forward pass
    loss = criterion(y_pred, Y) # Compute loss
    loss.backward()            # Backpropagation
    optimizer.step()           # Update weights

print("Training complete")
```

```python
pred = model(X)
print("Predicted energies:\n", pred)
print("True energies:\n", Y)
```
